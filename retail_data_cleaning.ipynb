```python
import pandas as pd
import numpy as np
```

```python
# loading dataset
df = pd.read_csv("online_retail.csv", encoding='ISO-8859-1')

# quick look at data
df.head()
```

```python
# shape of dataset
df.shape
# checking columns and missing values
df.info()
# basic stats
df.describe()
```

```python
# checking top countries
df['Country'].value_counts().head()
```

```python
# checking most frequent products
df['Description'].value_counts().head()
```

```python
# checking null values
df.isnull().sum()
```

```python
# removing rows where CustomerID is missing
df = df.dropna(subset=['CustomerID'])
# removing cancelled orders (InvoiceNo starts with 'C')
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
# keeping only positive quantity
df = df[df['Quantity'] > 0]
# keeping only positive price
df = df[df['UnitPrice'] > 0]
```

```python
df.shape
```

```python
df.head()
```

```python
# creating revenue column
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
# converting date column
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# extracting useful time features
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
# quick revenue check
df['Revenue'].sum()
```

```python
# revenue by month
df.groupby('Month')['Revenue'].sum()
```

```python
# saving cleaned data
df.to_csv("cleaned_retail.csv", index=False)
```
